```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Nov 21 14:54:19 2019

@author: shih3801
"""
import pandas as pd
import numpy as np
import glob
import matplotlib.pyplot as plt
import matplotlib
from Ecg_preprocess import Ecg_preprocess
import pixiedust

class Visualization():
    def __init__(self, path):
        self.path = path
        return


    def draw_ecg(self, data, times, pantomp_draw_peaks, hamilton_draw_peaks, engzee_draw_peaks, name):
        """Draw the two types of peaks on original ecg data
        data: original ecg data
        filter_ecg: used filter on the ecg data
        times: as the x coordinate
        pantomp_draw_peaks: use pan-tompkins algorithm to detect R-peaks; only draw 6 peaks
        hamilton_draw_peaks: use hamilton algorithm to detect R-peaks; only draw 6 peaks
        name: save the image based on the name
        three: draw one more peaks-engzee peaks
        """
        ymin = np.min(data)
        ymax = np.max(data)
        index = 0
        grid = (3, 1)
        fs = 200
        plt.subplot2grid(grid, (index, 0))
        # Calculate time values in seconds
        times = np.arange(data.shape[0], dtype='float') / fs
        plt.xlabel("Time (s)")
        plt.plot(times, data)
        plt.vlines([r / fs for r in pantomp_draw_peaks], ymin, ymax,
                   color="r",
                   linewidth=2)
        plt.title("Pantompkin R-peaks")
        plt.grid(True)

        index += 1
        plt.subplot2grid(grid, (index, 0))
        # Calculate time values in seconds
        plt.plot(times, data)
        plt.vlines([r / fs for r in hamilton_draw_peaks], ymin, ymax,
                   color="r",
                   linewidth=2)
        plt.title("Hamilton R-peaks ")
        plt.grid(True)


        index += 1
        plt.subplot2grid(grid, (index, 0))
        plt.plot(times, data)
        plt.vlines([r / fs for r in engzee_draw_peaks], ymin, ymax,
                   color="r",
                   linewidth=2)
        plt.title("Engzee R-peaks")
        plt.grid(True)
        plt.savefig('./jpg-ecg/' + name + '.jpg')

    def visual_draw(self, gsr=0, ecg=0, emg=0):
        """select gsr features in time frame with windows 50 and step 25,
        compute the max, min, mean, std of gsr in each window for five classes"""
        fs = 200  # corresponds to 60 beats per min (normal for human), assumed.
        lowcut = 0.05 * 3.3  # 9.9 beats per min
        highcut = 15  # 900 beats per min
        i = 0
        allFiles = glob.glob(self.path + "/*.csv")
        print('There is %d csv files in Part_A folder.' % (len(allFiles)))
        i = 0
        matplotlib.use("pgf")
        matplotlib.rcParams.update({
            "pgf.texsystem": "pdflatex",
            'font.family': 'serif',
            'text.usetex': True,
            'pgf.rcfonts': False,
        })
        for file in allFiles[0:len(allFiles)]:
            
            df = pd.read_csv(file, index_col=None, header=0, sep='\t')
            p = file.split("-")[0]
            user_name = p.split("/")[-1]
            if file.split('-')[1] == 'PA4':
                name = user_name + file.split("-")[1] + '-'
                # Save gsr as jpg
                if gsr:
                    data = df[['gsr']]
                    #name += str(-i)
                    numb = file.split('-')[2][:3]
                    print(numb)
                    name += numb
                    print('name is %s' %name)

                    #i += 1
                    df['time'] = np.arange(data.shape[0], dtype='float') / fs
                    df.plot(x='time', y='gsr')
                    plt.xlabel('Time')
                    plt.title('SCL')
                    plt.savefig('./jpg-gsr/' + name + '.pdf')
                if ecg:
                    df = df[['time', 'ecg']]
                    times = df['time'].values
                    data = df['ecg'].values
                    name += str(-i)
                    i += 1
                    df['time'] = np.arange(data.shape[0], dtype='float') / fs
                    df.plot(x='time', y='ecg')
                    plt.xlabel('Time')
                    plt.title('ECG')
                    plt.savefig('./jpg-ecg/' + name + '.pdf')

                if emg:
                    data = df[['emg_trapezius']]
                    name += str(-i)
                    i += 1
                    df['time'] = np.arange(data.shape[0], dtype='float') / fs
                    df.plot(x='time', y='emg_trapezius')  # kind='scatter', 
                    plt.xlabel('Time')
                    plt.title('EMG')
                    plt.savefig('./jpg-emg/' + name + '.pdf')
```
